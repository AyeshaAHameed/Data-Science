```python
x_list=['radius_mean', 'texture_mean', 'perimeter_mean','area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean','concave points_mean', 'symmetry_mean', 'fractal_dimension_mean','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se','compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se','fractal_dimension_se', 'radius_worst', 'texture_worst','perimeter_worst', 'area_worst', 'smoothness_worst','compactness_worst', 'concavity_worst', 'concave points_worst','symmetry_worst', 'fractal_dimension_worst']
```

```python
scaler=MinMaxScaler()
scaled=scaler.fit_transform(df[x_list])
```

```python
df[x_list] = scaled
```

```python
df.head()
```

```python
df=df.drop(['Unnamed: 32'],axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x = df.drop(['diagnosis','id'],axis=1)
y=df['diagnosis']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train , y_train)
```

```python
y_pred =model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print('acuracy : {:.2f}%'.format(accuracy*100))
```

```python
from sklearn.metrics import precision_score,recall_score
```

```python
precision = precision_score(y_test,y_train)
```
