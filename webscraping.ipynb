WEBSCRAPING BOOK DATA

```python
#import libraries
import requests 
from bs4 import BeautifulSoup
import pandas as pd

#send a request
url='https://books.toscrape.com/catalogue/category/books/science_22/index.html'#public url
response=requests.get(url)
response.status_code #200 means success

#parse the HTML
soup=BeautifulSoup(response.text,'lxml')

#extract data 
books=soup.find_all('article',class_='product_pod')
book_data=[]#empty list
#appending title,price and rating in book_data one by one
for book in books:
    title=book.h3.a['title']
    price=book.find('p',class_='price_color').text
    rating= book.p['class'][1]
    book_data.append([title,price,rating])
```

```python
#creating a dataframe
df = pd.DataFrame(book_data , columns=['Title','Price','Ratinf'])
 #save to CSV
df.to_csv('book_data.csv',index=False)

#display results
df.head()
```
